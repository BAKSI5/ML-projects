## 1. Define each of the following machine learning terms in your own words:

**i.**

**Training dataset**: This is like the practice material for a machine learning model. It's a set of examples that the model uses to learn and adjust its predictions.

**Testing dataset**: It is a separate set of examples that the model hasn't seen during training. We use it to see how well the model can make predictions on new, unseen data.

**Validation dataset**: It is a smaller set of data used to fine-tune the model's settings while training, to make sure it's learning correctly.

**ii.**

**Ground truth**: This is the correct answer or the right label for something in the data.

**Label**: This is like a tag or a name that we give to something in our data to describe or classify it. The information we want our machine learning model to predict or figure out. For example, in an email spam filter, the label might be "spam" or "not spam." In image recognition, the label could be "cat" or "dog." Labels help the model understand and categorize different things in the data.

**iii.**

**Pre-processing**: Preprocessing means getting your data ready for analysis or training a machine learning model. It involves various steps like cleaning, organizing, and transforming data so that it's in a suitable format for the model to work with. Preprocessing helps ensure that your data is in the best possible condition for the machine learning model to learn from and make accurate predictions or analyses.

**Feature**: These are the important parts of data that the model pays attention to when making predictions. In an image, features might be things like color or shape.

**Numerical**: Anything that's a number related, it can be counted or measured.

**Nominal**: This is something that doesn't have a numerical value, more like categories or labels. For example, colors like "red," "blue," and "green" are nominal.

**iv.**

**Decision surface**: This is like an imaginary line or boundary that the model draws to separate things into different groups.

**v.**

**Model validation**: This is the process of checking how well our machine learning model is performing. We use different tools like accuracy to measure this, testing data, cross-validation to test the model multiple times.
Model validation helps understand how reliable and effective our machine learning model is, and it's a crucial step in the development process to ensure the model can perform well on new, unseen data.

**Accuracy**: This tells you how many answers your model got right compared to the total number of answers.

**Cross-validation**: This is a way to test the model's performance multiple times with different parts of data.

**vi.**

**Parameters**: These are the settings inside the model that it learns from the training data.

**Hyperparameters**: These are settings that we choose before training. 

**Overfit**: Overfitting is when a model learns the training data too well, including the noise or random stuff, and doesn't work well with new, unseen data.

## 5. Using the scikit-learn class descriptions for Naive Bayes and decision trees, classify the Iris dataset in question 3. 
Your code should be very similar to that in the Module 1 Jupyter notebook. In classification with SVM section, the dataset is divided into two portions, one for training and the other for testing. Make sure you use the same input data for the Naive Bayes classifier, and the decision tree classifier.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load the dataset
iris = load_iris()
X, y = iris.data, iris.target

# Learn the data on the first half of the dataset - 50% data is used as the training set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Create and Train Naive Bayes classifier
Naive_Bayes_classifier = GaussianNB()
Naive_Bayes_classifier.fit(X_train, y_train)
Naive_Bayes_predictions = Naive_Bayes_classifier.predict(X_test)

# Create and Train Decision Tree classifier
Decision_Tree_classifier = DecisionTreeClassifier()
Decision_Tree_classifier.fit(X_train, y_train)
Decision_Tree_predictions = Decision_Tree_classifier.predict(X_test)

# Calculate accuracy
Naive_Bayes_acc = accuracy_score(y_test, Naive_Bayes_predictions)
Decision_Tree_acc = accuracy_score(y_test, Decision_Tree_predictions)

# Print classification report
print(f"Classification report for Naive Bayes:\nAccuracy={Naive_Bayes_acc:.3f}\n")
print(f"Classification report for Decision Tree:\nAccuracy={Decision_Tree_acc:.3f}\n")

# Print confusion matrices
print("Confusion matrix for Naive Bayes:")
print(confusion_matrix(y_test, Naive_Bayes_predictions))
print("\nConfusion matrix for Decision Tree:")
print(confusion_matrix(y_test, Decision_Tree_predictions))
```

Answer the following questions:

**i. Which classifier has the highest performance?**

The Naive Bayes classifier has the highest performance with an accuracy of 0.987, compared to the Decision Tree classifier's accuracy of 0.947. Therefore, the Naive Bayes classifier performs better in terms of accuracy on this dataset.

**iii. Will the performance plateau? Show it on a plot.**

Hints
 You need to repeat train-test with different samples to collect statistics (mean/stdev), 
e.g. 10 times for each experimental point. Plotting each experimental data point with 
mean and stdev helps reasoning, and answering how much training is sufficient.
 Use sklearn.model_selection.train_test_split and make sure the shuffle is on. 
Why

```python
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

# Define a range of training dataset sizes (e.g., 1%, 5%, 10%, ..., 90%)
training_sizes = np.arange(0.01, 0.91, 0.05)

# Initialize lists to store mean accuracies
Naive_Bayes_mean_acc = []
Naive_Bayes_std_acc = []
Decision_Tree_mean_acc = []
Decision_Tree_std_acc = []

# Repeat train-test 10 times for each experimental point
iterations = 10

for training_size in training_sizes:
    Naive_Bayes_accs = []
    Decision_Tree_accs = []
    for _ in range(iterations):
        # Split the data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=(1 - training_size), random_state=None, shuffle=True)

        # Train Naive Bayes classifier
        Naive_Bayes_classifier = GaussianNB()
        Naive_Bayes_classifier.fit(X_train, y_train)
        Naive_Bayes_predictions = Naive_Bayes_classifier.predict(X_test)
        Naive_Bayes_accuracy = accuracy_score(y_test, Naive_Bayes_predictions)
        Naive_Bayes_accs.append(Naive_Bayes_accuracy)

        # Train Decision Tree classifier
        Decision_Tree_classifier = DecisionTreeClassifier()
        Decision_Tree_classifier.fit(X_train, y_train)
        Decision_Tree_predictions = Decision_Tree_classifier.predict(X_test)
        Decision_Tree_accuracy = accuracy_score(y_test, Decision_Tree_predictions)
        Decision_Tree_accs.append(Decision_Tree_accuracy)

    # Calculate mean accuracy for each classifier at this training size
    Naive_Bayes_mean_acc.append(np.mean(Naive_Bayes_accs))
    Naive_Bayes_std_acc.append(np.std(Naive_Bayes_accs))
    Decision_Tree_mean_acc.append(np.mean(Decision_Tree_accs))
    Decision_Tree_std_acc.append(np.std(Decision_Tree_accs))

# Plot the results
plt.figure(figsize=(10, 6))
plt.errorbar(training_sizes * 100, Naive_Bayes_mean_acc, yerr=Naive_Bayes_std_acc, marker='o', label='Naive Bayes')
plt.errorbar(training_sizes * 100, Decision_Tree_mean_acc, yerr=Decision_Tree_std_acc, marker='x', label='Decision Tree')
plt.xlabel('Training Dataset Size (%)')
plt.ylabel('Mean Accuracy')
plt.title('Performance Plateau')
plt.legend()
plt.grid(True)
plt.show()
```
