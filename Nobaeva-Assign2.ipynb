**1. Perceptron:**

Perceptron is a simple linear classifier that separates two classes of data points using a hyperplane. It updates weights to minimize classification errors and applys a threshold function to make predictions but doesn't directly optimize a cost function. It's simple, efficient, and work well for linearly separable datasets but may not good fit on complex, non-linear problems.

*Optimization Problem*: The Perceptron seeks to find a hyperplane that separates two classes without directly minimizing a cost function; it updates weights to minimize classification errors.
*Speed*: Generally fast due to incremental weight updates.
*Strengths*: Effective for linearly separable data but less suitable for complex problems.
*Robustness*: Sensitive to initial conditions and noisy data; may not converge for non-linearly separable data.
*Feature Type*: Works well with numeric features.

**2. SVM (Support Vector Machine):**

A powerful and versatile classifier used for both classification and regression tasks. It finds the optimal hyperplane with the maximum margin between classes by minimizing a hinge loss function. SVMs handle high-dimensional data and complex decision boundaries. Effective for both linear and non-linear data, a wide range of applications, including image classification, text classification, and more.

*Optimization Problem*: SVM aims to find the hyperplane with the maximum margin between classes by minimizing the hinge loss, a cost function that penalizes misclassifications.
*Speed*: Slower than the Perceptron but efficient for non-linear data using the kernel trick.
*Strengths*: Strong in finding optimal separating hyperplanes, effective for non-linear data with kernels.
*Robustness*: Robust to outliers and works well with high-dimensional data.
*Feature Type*: Suitable for both numeric and categorical features.

**3. Decision Tree:**

A popular machine learning algorithm used for both classification and regression tasks. A tree-like structure where each internal node represents a feature or attribute, and each leaf node represents a class label or a numerical value. Decision Trees follows a path from the root node to a leaf node making decisions based on the values of input features. Suitable for non-linear relationships, can handle both categorical and numerical data, and often used for tasks like classification and data exploration.

*Optimization Problem*: Decision Trees split data recursively to maximize information gain or minimize impurity (e.g., Gini impurity or entropy).
*Speed*: Training is fast but can lead to overfitting; pruning can enhance speed and robustness.
*Strengths*: Can model non-linear relationships, handle mixed data types, and offer interpretability.
*Robustness*: Prone to overfitting, mitigated with pruning and Random Forests.
*Feature Type*: Compatible with both numeric and categorical features.

**4. Random Forest:**

An ensemble learning algorithm that improves predictive accuracy and reduces overfitting by combining multiple trees. It works by constructing a collection of decision trees, each trained on a random subset of the data and features. Random Forest offers robustness, ability to handle high-dimensional data, resistance to overfitting, and provides good predictive performance.

*Optimization Problem*: Random Forest is an ensemble of Decision Trees; each tree minimizes impurity or maximizes information gain.
*Speed*: Slower to train than individual Decision Trees but often faster than more complex models like SVM with large datasets.
*Strengths*: Combines multiple Decision Trees to reduce overfitting and enhance predictive performance, and it is robust to outliers and noise.
*Robustness*: Robust and adaptable to various data types and complexities.
*Feature Type*: Works effectively with both numeric and categorical features.

Which one to try first depends on the data and problem. For linearly separable data we can start with the Perceptron or SVM. For non-linear data, SVM with an appropriate kernel or a Decision Tree can be good options. If interpretability is crucial, we can begin with a Decision Tree. For robustness and better performance, we might consider Random Forest as it often provides a good balance between speed, strength, and robustness.

**Using online resources, research and find other classifier performance metrics which are also as common as the accuracy metric. Provide the mathematical equations for them and explain in your own words the meaning of the different metrics you found. Note

1. Precision (Positive Predictive Value):

Mathematical Equation: Precision = TruePositive / (TruePositive + FalsePositive)
<br>
Precision assesses the fraction of accurate positive predictions (positively identified cases) relative to the total number of positive predictions generated by the model. It provides insight into the query: "Out of the model's positive predictions, what proportion were accurate?"

2. Recall (Sensitivity, True Positive Rate):

Mathematical Equation: Recall = TruePositive / (TruePositive + FalseNegative)
<br>
Recall measures the ratio of correctly identified positive cases (true positives) to the total number of actual positive cases. It answers the inquiry: "Among all the real positive cases, how many did the model successfully detect?"

3. F1-Score:

Mathematical Equation: F1-Score = 2 * (Precision * Recall) / (Precision + Recall)
<br>
The F1-Score represents the harmonic average of precision and recall, serving as an equilibrium point between these two measures, particularly valuable in cases of imbalanced class distribution. It offers a unified metric that consolidates both precision and recall into a single value.

4. Specificity (True Negative Rate):

Mathematical Equation: Specificity = TrueNegative / (TrueNegative + FalsePositive)
<br>
Specificity quantifies the ratio of true negative predictions (accurately identified negative cases) relative to the total number of actual negative cases. It addresses the query: "Among all the genuine negative cases, how many did the model correctly classify as negative?"

5. False Positive Rate (FPR):

Mathematical Equation: FPR = FalsePositive / (FalsePositive + TrueNegative)
<br>
FPR calculates the ratio of false positive predictions in comparison to all actual negative cases. It serves as the counterpart to specificity and aids in evaluating the model's capacity to minimize false alarms.

6. True Negative Rate (TNR) (Specificity):

Mathematical Equation: TNR = TrueNegative / (TrueNegative + FalsePositive)
<br>
TNR signifies the proportion of true negative predictions relative to all the genuine negative cases.

7. Area Under the Receiver Operating Characteristic Curve (AUC-ROC):

Mathematical Equation: AUC-ROC is calculated by plotting the ROC curve (Receiver Operating Characteristic) and measuring the area under the curve. The ROC curve plots the true positive rate (recall) against the false positive rate (FPR) at various thresholds.
<br.
AUC-ROC evaluates the model's capacity to distinguish between positive and negative classes while varying the threshold levels. A higher AUC-ROC value signifies superior discriminatory performance.

8. Area Under the Precision-Recall Curve (AUC-PR):

Mathematical Equation: AUC-PR is calculated by plotting the precision against recall at various thresholds.
<br>
AUC-PR offers a concise representation of how a model manages the trade-off between precision and recall. This metric is especially valuable in scenarios involving imbalanced datasets where one class is significantly less frequent than the other.

```python
# Drop the 'Serial No.' column as we don't need it
df = df.drop(columns=['Serial No.'])

# Calculate the correlation matrix manually
correlation_matrix = pd.DataFrame(index=df.columns, columns=df.columns)

for col1 in df.columns:
    for col2 in df.columns:
        # Calculate the correlation between col1 and col2 using Pearson's correlation formula
        mean_col1 = df[col1].sum() / len(df)
        mean_col2 = df[col2].sum() / len(df)
        numerator = ((df[col1] - mean_col1) * (df[col2] - mean_col2)).sum()
        denominator_col1 = ((df[col1] - mean_col1) ** 2).sum()
        denominator_col2 = ((df[col2] - mean_col2) ** 2).sum()
        correlation = numerator / (denominator_col1 ** 0.5 * denominator_col2 ** 0.5)
        
        # Store the correlation in the matrix
        correlation_matrix.loc[col1, col2] = correlation

# Fill the diagonal with 1's
for col in df.columns:
    correlation_matrix.loc[col, col] = 1.0

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Create pair plots
sns.pairplot(df)
plt.title("Pair Plot of Features")
plt.show()
```

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Calculate the correlation matrix using the selected numeric columns
correlation_matrix = numeric_columns.corr()

# For a visual aspect create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

We shouldn't use the 'Serial No.' column in our analysis as it's just an identifier and doesn't provide any meaningful information.

The diagonal of the correlation matrix should have all 1's because it represents the correlation of each feature with itself, which is perfect correlation. It is very to easy to see it on the heatmap above.

When looking at the correlations between variables, it's essential to consider the correlation of each feature with the target variable ('Chance of Admit'). Features with higher absolute correlations (positive or negative) with the target variable are generally more important for prediction. 

As we can see from the heatmap the high correlation (0.88) between CGPA and the "Chance of Admit" in the dataset indicates that students with higher CGPA scores are more likely to be admitted to graduate programs. This strong relationship makes sense because universities often consider academic performance when making admission decisions.
